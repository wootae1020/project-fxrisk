### 경제불확실성지수 (Economic Policy Uncertainty Index)

- **경제불확실성(Economic Policy Uncertainty, EPU)** 지수는 실시간으로 생성되는 뉴스 기사의 **텍스트 데이터를 분석**해 경제 흐름을 파악하는 지표이며, **매월 경제불확실성 지수를 제공**함.

- EPU 지수는 **언론사**의 전체 기사 중에서  
  **경제(Economy)**, **정책(Policy)**, **불확실성(Uncertainty)** 과 관련된 **키워드의 비중을 계산**하여 시계열 그래프로 시각화함.

- 경제불확실성 : 미래의 경제 상황을 예측할 수 없는 상태
- 미국 대통령 선거, 일본 수출 규제, 코로나 팬데믹 등의 큰 사건이 발생했을 때, 앞으로의 경제 상황을 예측하기 매우 어려워짐

- 경제 불확실성은 직접 관측이 불가능하여 다양한 지표를 통해 간접적으로 가늠

- EPU지수 : 미국 노스웨스턴대학교 교수 Baker 등이 개발한 지표로 뉴스 기사에서 경제(E), 정책(P), 불확실성(U) 관련 단어 출현 빈도를 표준화하고 정규화하여 수치화한 것

- 실시간으로 생성되는 뉴스데이터를 분석하여 산출하기 때문에 경제 상황을 준 실시간으로 모니터링 가능

### 기존 EPU 지수와의 차별성 :

- 1. 4가지 방법으로 단어를 확대해서 서로 다른 한국 EPU를 비교

- 2. EPU 개발에 사용된 데이터를 활용해 월별 이슈 토픽모델링해 수치를 정성적으로 해석

#### 사용된 언론사 목록
- 경향신문, 국민일보, 동아일보, 문화일보, 서울신문, 세계일보, 중앙일보, 조선일보, 한겨례, 매일경제, 서울경제, 한국경제


# 처리 순서
- 언론사 n개 정의
- 경제, 불확실성, 정책 카테고리에 각각 하나의 용어를 공동으로 포함하는 기사를 수집
- 기간에 맞게 뉴스 기사에 대한 데이터를 수집
- 신문사, 월별로 관련 기사 수를 집계 후, 동일 신문의 총 기사 수로 집계하여 상대 빈도 수를 얻음
- 각 신문의 상대 EPU 수를 전체 표본 기간의 단위 표준 편차로 표준화
- 신문의 표준화 시리즈를 월별로 평균하여 전체 월간 한국 EPU지수
- EPU 디수를 표본 기간 중 평균 100으로 정규화
