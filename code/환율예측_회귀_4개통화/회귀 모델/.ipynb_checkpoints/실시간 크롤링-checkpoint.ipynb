```python
def EURKRW():
    df = pd.DataFrame(columns=["Date", "Close", "Open", "High", "Low", "유로환율_변동%_x"])
    
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    service = Service()
    
    driver = webdriver.Chrome(service=service, options=options)
    url = "https://kr.investing.com/currencies/eur-krw-historical-data"
    driver.get(url)
    time.sleep(5)
    
    date_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[1]'
    close_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[2]'
    open_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[3]'
    high_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[4]'
    low_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[5]'
    volume_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[6]'
    change_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[7]'
    
    date = driver.find_element(By.XPATH, date_xpath).text
    close = driver.find_element(By.XPATH, close_xpath).text
    open_ = driver.find_element(By.XPATH, open_xpath).text
    high = driver.find_element(By.XPATH, high_xpath).text
    low = driver.find_element(By.XPATH, low_xpath).text
    volume = driver.find_element(By.XPATH, volume_xpath).text
    change = driver.find_element(By.XPATH, change_xpath).text
    
    def clean_number(text):
        return float(text.replace(',', '').replace('%', ''))
    
    from datetime import datetime
    date = datetime.strptime(date.replace('월', '-'), '%m- %d, %Y').date()
    
    new_row = {
        "Date": date,
        "Close": clean_number(close),
        "Open": clean_number(open_),
        "High": clean_number(high),
        "Low": clean_number(low),
        "유로환율_변동%_x": clean_number(change)
    }
    
    if date not in df["Date"].values:
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    
    df = df.sort_values("Date").reset_index(drop=True)
    
    driver.quit()
    return df

def DAX():
    df = pd.DataFrame(columns=["Date", "Dax_종가", "Dax_시가", "Dax_고가", "Dax_저가", "Dax_거래량(백 만)", "Dax_변동"])
    
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    service = Service()
    
    driver = webdriver.Chrome(service=service, options=options)
    url = "https://kr.investing.com/indices/germany-30-historical-data"
    driver.get(url)
    time.sleep(5)
    
    date_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[1]'
    close_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[2]'
    open_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[3]'
    high_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[4]'
    low_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[5]'
    volume_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[6]'
    change_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[7]'
    
    date = driver.find_element(By.XPATH, date_xpath).text
    close = driver.find_element(By.XPATH, close_xpath).text
    open_ = driver.find_element(By.XPATH, open_xpath).text
    high = driver.find_element(By.XPATH, high_xpath).text
    low = driver.find_element(By.XPATH, low_xpath).text
    volume = driver.find_element(By.XPATH, volume_xpath).text
    change = driver.find_element(By.XPATH, change_xpath).text
    
    def clean_number(text):
        return float(text.replace(',', '').replace('M', '').replace('%', ''))
    
    from datetime import datetime
    date = datetime.strptime(date.replace('월', '-'), '%m- %d, %Y').date()
    
    new_row = {
        "Date": date,
        "Dax_종가": clean_number(close),
        "Dax_시가": clean_number(open_),
        "Dax_고가": clean_number(high),
        "Dax_저가": clean_number(low),
        "Dax_거래량(백 만)": clean_number(volume),
        "Dax_변동%": clean_number(change)
    }
    
    if date not in df["Date"].values:
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    
    df = df.sort_values("Date").reset_index(drop=True)
    
    driver.quit()
    return df

def VSTOXX():
    df = pd.DataFrame(columns=["Date", "VSTOXX_종가", "VSTOXX_시가", "VSTOXX_고가", "VSTOXX_저가", "VSTOXX_변동%"])
    
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    service = Service()
    
    driver = webdriver.Chrome(service=service, options=options)
    url = "https://kr.investing.com/indices/stoxx-50-volatility-vstoxx-eur-historical-data"
    driver.get(url)
    time.sleep(5)
    
    date_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[1]'
    close_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[2]'
    open_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[3]'
    high_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[4]'
    low_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[5]'
    volume_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[6]'
    change_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[7]'
    
    date = driver.find_element(By.XPATH, date_xpath).text
    close = driver.find_element(By.XPATH, close_xpath).text
    open_ = driver.find_element(By.XPATH, open_xpath).text
    high = driver.find_element(By.XPATH, high_xpath).text
    low = driver.find_element(By.XPATH, low_xpath).text
    volume = driver.find_element(By.XPATH, volume_xpath).text
    change = driver.find_element(By.XPATH, change_xpath).text
    
    def clean_number(text):
        return float(text.replace(',', '').replace('M', '').replace('%', ''))
    
    from datetime import datetime
    date = datetime.strptime(date.replace('월', '-'), '%m- %d, %Y').date()
    
    new_row = {
        "Date": date,
        "VSTOXX_종가": clean_number(close),
        "VSTOXX_시가": clean_number(open_),
        "VSTOXX_고가": clean_number(high),
        "VSTOXX_저가": clean_number(low),
        "VSTOXX_변동%": clean_number(change)
    }
    
    if date not in df["Date"].values:
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    
    df = df.sort_values("Date").reset_index(drop=True)
    
    driver.quit()
    return df

def EUROSTOXX50():
    df = pd.DataFrame(columns=["Date", "EUROSTOXX50_종가", "EUROSTOXX50_시가", "EUROSTOXX50_고가", "EUROSTOXX50_저가", "EUROSTOXX50_변동%"])
    
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    service = Service()
    
    driver = webdriver.Chrome(service=service, options=options)
    url = "https://kr.investing.com/indices/germany-30-historical-data"
    driver.get(url)
    time.sleep(5)
    
    date_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[1]'
    close_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[2]'
    open_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[3]'
    high_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[4]'
    low_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[5]'
    volume_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[6]'
    change_xpath = '//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[7]'
    
    date = driver.find_element(By.XPATH, date_xpath).text
    close = driver.find_element(By.XPATH, close_xpath).text
    open_ = driver.find_element(By.XPATH, open_xpath).text
    high = driver.find_element(By.XPATH, high_xpath).text
    low = driver.find_element(By.XPATH, low_xpath).text
    volume = driver.find_element(By.XPATH, volume_xpath).text
    change = driver.find_element(By.XPATH, change_xpath).text
    
    def clean_number(text):
        return float(text.replace(',', '').replace('M', '').replace('%', ''))
    
    from datetime import datetime
    date = datetime.strptime(date.replace('월', '-'), '%m- %d, %Y').date()
    
    new_row = {
        "Date": date,
        "EUROSTOXX50_종가": clean_number(close),
        "EUROSTOXX50_시가": clean_number(open_),
        "EUROSTOXX50_고가": clean_number(high),
        "EUROSTOXX50_저가": clean_number(low),
        "EUROSTOXX50_변동": clean_number(change)
    }
    
    if date not in df["Date"].values:
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    
    df = df.sort_values("Date").reset_index(drop=True)
    
    driver.quit()
    return df
```

```python
import pandas as pd
import time
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

class WebDataFetcher:
    def __init__(self):
        self.options = webdriver.ChromeOptions()
        self.options.add_argument("--headless")
        self.options.add_argument("--no-sandbox")
        self.options.add_argument("--disable-dev-shm-usage")
        self.service = Service()

    def _clean_number(self, text):
        return float(text.replace(',', '').replace('%', '').replace('M', ''))

    def _get_driver(self):
        return webdriver.Chrome(service=self.service, options=self.options)

    def _fetch_common(self, url):
        driver = self._get_driver()
        driver.get(url)
        time.sleep(5)

        get = lambda xpath: driver.find_element(By.XPATH, xpath).text
        try:
            date = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[1]')
            close = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[2]')
            open_ = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[3]')
            high = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[4]')
            low = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[5]')
            volume = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[6]')
            change = get('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[2]/div[3]/table/tbody/tr[1]/td[7]')
        finally:
            driver.quit()

        date = datetime.strptime(date.replace("월", "-"), '%m- %d, %Y').date()
        return date, close, open_, high, low, volume, change

    def fetch_EURKRW(self):
        url = "https://kr.investing.com/currencies/eur-krw-historical-data"
        date, close, open_, high, low, _, change = self._fetch_common(url)
        return pd.DataFrame([{
            "Date": date,
            "Close": self._clean_number(close),
            "Open": self._clean_number(open_),
            "High": self._clean_number(high),
            "Low": self._clean_number(low),
            "유로환율_변동%_x": self._clean_number(change)
        }])

    def fetch_DAX(self):
        url = "https://kr.investing.com/indices/germany-30-historical-data"
        date, close, open_, high, low, volume, change = self._fetch_common(url)
        return pd.DataFrame([{
            "Date": date,
            "Dax_종가": self._clean_number(close),
            "Dax_시가": self._clean_number(open_),
            "Dax_고가": self._clean_number(high),
            "Dax_저가": self._clean_number(low),
            "Dax_거래량(백 만)": self._clean_number(volume),
            "Dax_변동%": self._clean_number(change)
        }])

    def fetch_VSTOXX(self):
        url = "https://kr.investing.com/indices/stoxx-50-volatility-vstoxx-eur-historical-data"
        date, close, open_, high, low, _, change = self._fetch_common(url)
        return pd.DataFrame([{
            "Date": date,
            "VSTOXX_종가": self._clean_number(close),
            "VSTOXX_시가": self._clean_number(open_),
            "VSTOXX_고가": self._clean_number(high),
            "VSTOXX_저가": self._clean_number(low),
            "VSTOXX_변동%": self._clean_number(change)
        }])

    def fetch_EUROSTOXX50(self):
        url = "https://kr.investing.com/indices/eu-stoxx50-historical-data"
        date, close, open_, high, low, _, change = self._fetch_common(url)
        return pd.DataFrame([{
            "Date": date,
            "EUROSTOXX50_종가": self._clean_number(close),
            "EUROSTOXX50_시가": self._clean_number(open_),
            "EUROSTOXX50_고가": self._clean_number(high),
            "EUROSTOXX50_저가": self._clean_number(low),
            "EUROSTOXX50_변동%": self._clean_number(change)
        }])
```

```python
import yfinance as yf
import pandas as pd

# 티커 목록
tickers = {
    "DAX": "^GDAXI",
    "EUROSTOXX50": "^STOXX50E",
    "CAC40": "^FCHI",
    "KRW_per_EUR": "EURKRW=X"
}

# 데이터프레임들을 저장할 리스트
df_list = []

for name, symbol in tickers.items():
    ticker = yf.Ticker(symbol)
    hist = ticker.history(period="2d", interval="1d")  # 전일 종가 포함해서 2일치
    if len(hist) >= 2:
        hist = hist.copy()
        hist = hist.tail(1)  # 오늘자 1일만 사용
        hist = hist[["Open", "High", "Low", "Close", "Volume"]]
        hist.columns = [f"{name}_{col}" for col in hist.columns]  # 컬럼 이름 변경
        hist["Date"] = hist.index.date
        hist.reset_index(drop=True, inplace=True)
        df_list.append(hist)
    else:
        print(f"{name} 데이터 부족")

# 여러 데이터프레임을 날짜 기준으로 병합
df_merged = df_list[0]
for df in df_list[1:]:
    df_merged = pd.merge(df_merged, df, on="Date", how="outer")

# 변동량 추가
for name in tickers.keys():
    close_col = f"{name}_Close"
    prev_day = yf.Ticker(tickers[name]).history(period="2d", interval="1d")
    if len(prev_day) == 2:
        change = prev_day["Close"].iloc[-1] - prev_day["Close"].iloc[-2]
        df_merged[f"{name}_Change"] = change
```

```python
import yfinance as yf
import pandas as pd

# 유럽 주요 지수만
tickers = {
    "DAX": "^GDAXI",
    "EUROSTOXX50": "^STOXX50E",
    "CAC40": "^FCHI"
}

df_list = []

# 20일치 데이터 수집
for name, symbol in tickers.items():
    ticker = yf.Ticker(symbol)
    hist = ticker.history(period="21d", interval="1d")  # 오늘 포함 20일 확보

    if len(hist) >= 2:
        hist = hist.copy()
        hist = hist[["Open", "High", "Low", "Close", "Volume"]]
        hist.columns = [f"{name}_{col}" for col in hist.columns]
        hist["Date"] = hist.index.date
        hist.reset_index(drop=True, inplace=True)
        df_list.append(hist)
    else:
        print(f"{name} 데이터 부족")

# 병합
df_merged = df_list[0]
for df in df_list[1:]:
    df_merged = pd.merge(df_merged, df, on="Date", how="outer")

# 정렬
df_merged = df_merged.sort_values("Date").reset_index(drop=True)

# 각 종목별 변동량 계산
for name in tickers.keys():
    close_col = f"{name}_Close"
    change_col = f"{name}_Change"
    if close_col in df_merged.columns:
        df_merged[change_col] = df_merged[close_col].diff()

print("유럽 지수 데이터")
print(df_merged.tail())
```
