```python
import pandas as pd
import numpy as np
```

```python
merged_data=pd.read_csv('merged_data.csv')
```

```python
merged_data
```

```python
merged_data=merged_data.drop(['상승','변화율'],axis=1)
```

```python
merged_data['변화율1']=merged_data['원/달러(종가 15:30)'].pct_change(periods=1).shift(-1)
merged_data['상승1']=(merged_data['변화율1']>0).astype(int)
merged_data['변화율2']=merged_data['원/달러(종가 15:30)'].pct_change(periods=2).shift(-2)
merged_data['상승2']=(merged_data['변화율2']>0).astype(int)
merged_data['변화율3']=merged_data['원/달러(종가 15:30)'].pct_change(periods=3).shift(-3)
merged_data['상승3']=(merged_data['변화율3']>0).astype(int)
merged_data['변화율4']=merged_data['원/달러(종가 15:30)'].pct_change(periods=4).shift(-4)
merged_data['상승4']=(merged_data['변화율4']>0).astype(int)
merged_data['변화율5']=merged_data['원/달러(종가 15:30)'].pct_change(periods=5).shift(-5)
merged_data['상승5']=(merged_data['변화율5']>0).astype(int)
merged_data['변화율6']=merged_data['원/달러(종가 15:30)'].pct_change(periods=6).shift(-6)
merged_data['상승6']=(merged_data['변화율6']>0).astype(int)
merged_data['변화율7']=merged_data['원/달러(종가 15:30)'].pct_change(periods=7).shift(-7)
merged_data['상승7']=(merged_data['변화율7']>0).astype(int)
merged_data['변화율8']=merged_data['원/달러(종가 15:30)'].pct_change(periods=8).shift(-8)
merged_data['상승8']=(merged_data['변화율8']>0).astype(int)
```

```python
merged_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x=merged_data[['sentiment']].values
y=merged_data[['상승1','상승2','상승3','상승4','상승5','상승6','상승7','상승8']].values
```

```python
tr_x,tt_x,tr_y,tt_y=train_test_split(x,y,train_size=0.8,shuffle=False)
print(tr_x.shape,tt_x.shape,tr_y.shape,tt_y.shape)
```

```python
model1=RandomForestClassifier().fit(tr_x,tr_y[:,0])
model2=RandomForestClassifier().fit(tr_x,tr_y[:,1])
model3=RandomForestClassifier().fit(tr_x,tr_y[:,2])
model4=RandomForestClassifier().fit(tr_x,tr_y[:,3])
model5=RandomForestClassifier().fit(tr_x,tr_y[:,4])
model6=RandomForestClassifier().fit(tr_x,tr_y[:,5])
model7=RandomForestClassifier().fit(tr_x,tr_y[:,6])
model8=RandomForestClassifier().fit(tr_x,tr_y[:,7])
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
pred_y1=model1.predict(tt_x)
pred_y2=model2.predict(tt_x)
pred_y3=model3.predict(tt_x)
pred_y4=model4.predict(tt_x)
pred_y5=model5.predict(tt_x)
pred_y6=model6.predict(tt_x)
pred_y7=model7.predict(tt_x)
pred_y8=model8.predict(tt_x)
```

```python
print(accuracy_score(tt_y[:,0],pred_y1))
print(f1_score(tt_y[:,0],pred_y1))
print(accuracy_score(tt_y[:,1],pred_y2))
print(f1_score(tt_y[:,1],pred_y2))
print(accuracy_score(tt_y[:,2],pred_y3))
print(f1_score(tt_y[:,2],pred_y3))
print(accuracy_score(tt_y[:,3],pred_y4))
print(f1_score(tt_y[:,3],pred_y4))
print(accuracy_score(tt_y[:,4],pred_y5))
print(f1_score(tt_y[:,4],pred_y5))
print(accuracy_score(tt_y[:,5],pred_y6))
print(f1_score(tt_y[:,5],pred_y6))
print(accuracy_score(tt_y[:,6],pred_y7))
print(f1_score(tt_y[:,6],pred_y7))
print(accuracy_score(tt_y[:,7],pred_y8))
print(f1_score(tt_y[:,7],pred_y8))
```

```python
news_data=pd.read_csv('news_data.csv',encoding='utf-8',index_col=0)
```

```python
news_data['일자']=pd.to_datetime(news_data['일자'],format="%Y%m%d")
```

```python
news_data[news_data.일자.isin(merged_data[merged_data.변화율1이상치==-1].날짜)]
```

```python
increase_group=pd.DataFrame({'일자':merged_data[(merged_data.변화율1이상치==-1)&(merged_data.상승==1)].날짜}).reset_index().drop('index',axis=1)
decrease_group=pd.DataFrame({'일자':merged_data[(merged_data.변화율1이상치==-1)&(merged_data.상승==0)].날짜}).reset_index().drop('index',axis=1)
steady_group=pd.DataFrame({'일자':merged_data[merged_data.변화율1이상치==1].날짜}).reset_index().drop('index',axis=1)
```

```python
increase_group=pd.merge(increase_group,news_data[['일자','키워드']],on='일자',how='inner')
decrease_group=pd.merge(decrease_group,news_data[['일자','키워드']],on='일자',how='inner')
steady_group=pd.merge(steady_group,news_data[['일자','키워드']],on='일자',how='inner')
```

```python
increase_group['키워드']=increase_group['키워드'].fillna('').apply(lambda x: x.replace(',',' '))
increase_group_daily=increase_group.groupby(['일자'])['키워드'].apply(lambda x: ' '.join(x)).reset_index()
decrease_group['키워드']=decrease_group['키워드'].fillna('').apply(lambda x: x.replace(',',' '))
decrease_group_daily=decrease_group.groupby(['일자'])['키워드'].apply(lambda x: ' '.join(x)).reset_index()
steady_group['키워드']=steady_group['키워드'].fillna('').apply(lambda x: x.replace(',',' '))
steady_group_daily=steady_group.groupby(['일자'])['키워드'].apply(lambda x: ' '.join(x)).reset_index()
```
